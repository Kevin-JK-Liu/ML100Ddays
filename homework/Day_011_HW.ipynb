```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
app_train = pd.read_csv('application_train.csv')
app_train.head()
```

```python
# 如果欄位中有 NA, describe 會有問題
app_train['AMT_ANNUITY'].describe()
```

```python
five_num = range(0,101)
for i in five_num:
    quantile_5s = np.percentile(app_train[~app_train['AMT_ANNUITY'].isnull()]['AMT_ANNUITY'], i)
    print('第{}分位數:'.format(i),quantile_5s)
```

```python
# Ignore NA, 計算五值
five_num = range(0,101)
quantile_5s = [np.percentile(app_train[~app_train['AMT_ANNUITY'].isnull()]['AMT_ANNUITY'], q = i) for i in five_num]
print(quantile_5s)
```
